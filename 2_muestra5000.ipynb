```python
# 2. Cargar archivo de tweets scrapeado
tweets_scrape_path = '/content/drive/MyDrive/TFM/                      '
tweets_df = pd.read_csv(tweets_scrape_path)

# Limpiar duplicados por tweet_id
tweets_df = tweets_df.drop_duplicates(subset=["tweet_id"])

# Filtrar para quedarnos solo con los tweet_id seleccionados
tweets_df = tweets_df[tweets_df["tweet_id"].astype(str).isin(muestra_ids)]

# Guardar archivo limpio
tweets_df.to_csv("tweets_scrape_limpio.csv", index=False)

# 3. Extraer fecha desde tweet_text
def extraer_fecha(texto):
    patron = r"\d{1,2}:\d{2} [AP]M ¬∑ [A-Za-z]{3,9} \d{1,2}, \d{4}"
    match = re.search(patron, texto)
    return match.group(0) if match else None

# Aplicar la funci√≥n
tweets_df["fecha_raw"] = tweets_df["tweet_text"].apply(extraer_fecha)
tweets_df["fecha_datetime"] = pd.to_datetime(tweets_df["fecha_raw"], errors='coerce')

# 4. An√°lisis temporal
print("üîç Intervalo temporal:")
print("Fecha m√≠nima:", tweets_df["fecha_datetime"].min())
print("Fecha m√°xima:", tweets_df["fecha_datetime"].max())

# Distribuci√≥n por mes
tweets_df["mes"] = tweets_df["fecha_datetime"].dt.to_period("M")
conteo_mensual = tweets_df["mes"].value_counts().sort_index()

print("\nüìà Tweets por mes:")
print(conteo_mensual)
```
